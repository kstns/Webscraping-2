##### This function captures information on multiple job openings from one page.
Selenium and XML language allows to scrape information from rendered HTML dynamic pages which most of the websites use. One page would have information on 50 positions. Each position shows: title, company, location, sometimes salary and short description. It would throw an exception in case requested XML element can't be found. The exception would cause None value to be written in the data.

```python
def scrape_one_page(df):
    scrape = driver.find_elements_by_xpath('//div[@class="job_seen_beacon"]')
    one_page_data = {}

    for i in range(len(scrape)):
        dct = {}
        
        try:
            dct['title'] = scrape[i].find_element_by_xpath('.//h2[contains(@class, "jobTitle")]//child::span[contains(@title, " ")]').text
        except:
            dct['title'] = None
        
        try:
            dct['company'] = scrape[i].find_element_by_xpath('.//child::span[@class="companyName"]').text
        except:
            dct['company'] = None
            
        try:
            dct['location'] = scrape[i].find_element_by_xpath('.//child::div[@class="companyLocation"]').text
        except:
            dct['location'] = None
        
        try:
            dct['salary'] = scrape[i].find_element_by_xpath('.//child::div[@class="metadata salary-snippet-container"]//child::div').text
        except:
            dct['salary'] = None
        
        dct['description'] = []
        try:
            for element in scrape[i].find_elements_by_xpath('.//tr[@class = "underShelfFooter"]//child::div[@class = "job-snippet"]//child::li'):
                dct['description'].append(element.text)
        except:
            dct['description'] = None

        one_page_data[i] = dct
    
    return pd.concat([df, pd.DataFrame.from_dict(one_page_data, orient='index')])
```

The function clicks "next page" button. Returns 1 if the action was performed and 0 otherwise.

```python
def next_page():
    try:
        driver.find_element_by_xpath('//a[@aria-label = "Next"]').click()
    except:
        return 0
    else:
        return 1
```

To close prompt window that opens from time to time on indeed.com

```python
def close_prompt_window():
    try:
        driver.find_element_by_xpath('//button[@autofocus=""]').click()
    except:
        return 0
    else:
        return 1
```

Saving pandas dataframe with captured information in a file on local disc. I can use the saved data anytime without re-scraping indeed.com

```python
def save_results(df, file_name):
    with open(file_name, 'wb') as f:
        pickle.dump(df, f, protocol=pickle.HIGHEST_PROTOCOL)
```

##### Main function to start scraping process.
This function stops when "next button" can't be identified on a webpage or it completed scraping process for 20 pages. Indeed.com only allows to see ~10 pages of job search results although there could be much more jobs found. That's why i decided to limit the scraping process by 20 iterations.

```python
def scrape(url, file_name):
    driver.get(url)
    sleep(3)
    
    df = pd.DataFrame(columns=['title', 'company', 'location', 'salary', 'description'])
    for i in range(20):
        sleep(1)
        #check_page_load()
        df = scrape_one_page(df)
        if next_page() == 0:
            break
        else:
            sleep(1)
            close_prompt_window()

    df.reset_index(drop=True, inplace=True)
    save_results(df, file_name)
    return df
```

To process salary information. Usually it is presented as a range - min and max. Ocasionally as one number. Sometimes as hourly rate which needs to be converted to annual number.

```python
def salary(x):
    salary = {}
    numbers = []
    
    if pd.isna(x):
        salary = None
    else:
        if x.lower().find('year') >= 0:
            salary['type'] = 'yearly'
        elif x.lower().find('hour') >= 0:
            salary['type'] = 'hourly'
        else:
            salary['type'] = 'other'

        numbers = [float(s.replace('$','').replace(',','')) for s in x.split() if s.find('$') >= 0]
        if len(numbers) == 2:
            salary['min'] = numbers[0]
            salary['mid'] = None
            salary['max'] = numbers[1]
        elif len(numbers) == 1:
            if x.lower().find('from') >= 0:
                salary['min'] = numbers[0]
                salary['mid'] = None
                salary['max'] = None
            elif x.lower().find('up to') >= 0:
                salary['min'] = None
                salary['mid'] = None
                salary['max'] = numbers[0]
            else:
                salary['min'] = None
                salary['mid'] = numbers[0]
                salary['max'] = None
        else:
            salary['min'] = None
            salary['mid'] = None
            salary['max'] = None
    
    return pd.Series(salary)
```

```python
df_salary = df['salary'].apply(salary)
df_salary.head()
```

The function to extract state and remote information from location.

```python
def state(x):
    location = x.split('\n')[0]
    if location.lower().find('location') >= 0 or location.lower().find('remote') >= 0:
        state = None
    elif location.find(',') == -1:
        state = location
    elif location.find(',') >= 0:
        state = location[location.find(',') + 2:location.find(',') + 4]
    else:
        state = 'Other'
    return state
```

#### This part is an attempt to do common words and phrases text analysis found in the description field

```python
import nltk
nltk.download('punkt')
nltk.download('stopwords')
from nltk.stem import PorterStemmer
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.util import ngrams
from collections import Counter

stop_words = set(stopwords.words('english'))
```

<br>
<br>

###### Removing punctuation, "stop words" such as articles, conjunctions, propositions.

```python
def text_process(x):
    words_only_list = re.sub("[^a-zA-Z]", " ", x).lower().split()
    no_sw_list = [w for w in words_only_list if not w in stop_words]
    stemm_words_list = [PorterStemmer().stem(w) for w in no_sw_list]
    return " ".join(stemm_words_list)
```

<br>
<br>

###### Original decsription field is a list of strings so converting it to a string here and applying function just defined in the previous step.

```python
df['description2'] = df.description.apply(lambda x: "".join(x))
```

```python
df['descr_text'] = df.description2.apply(text_process)
```
